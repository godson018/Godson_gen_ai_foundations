```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("/content/Reviews.csv", on_bad_lines='skip', engine='python')

print(df.head())
print(df.shape)
```

```python
light_reviews = df[df['Text'].str.contains('light', case=False, na=False)]

print("Number of reviews containing 'light':", len(light_reviews))
```

```python
for review in light_reviews['Text'].head(20):
    print(review)
    print("-"*100)
```

```python
sentence1 = "This olive oil is wonderful and very light on the stomach."

sentence2 = "The packaging was broken and let too much light inside the box."

sentence3 = "I prefer this brand because it is a light and healthy snack."
```

```python
sentences = [sentence1, sentence2, sentence3]
```

```python
vectorizer = CountVectorizer()

dtm = vectorizer.fit_transform(sentences)
```

```python
dense_matrix = dtm.toarray()

print(dense_matrix)
```

```python
print(vectorizer.get_feature_names_out())
```

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print("Vector 1:", vec_1)
print("Vector 2:", vec_2)
print("Vector 3:", vec_3)
```

```python
similarity_1_2 = np.dot(vec_1, vec_2)
```

```python
similarity_1_3 = np.dot(vec_1, vec_3)
```

```python
print("Similarity between Sentence 1 and Sentence 2:")
print(similarity_1_2)

print()

print("Similarity between Sentence 1 and Sentence 3:")
print(similarity_1_3)
```
